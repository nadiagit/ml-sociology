Which of the operators +, -, *, and / do something useful with the lists [1,10, 100] and [5, 4, 7]?

What happens if you apply the operators +, -, *, / to a list and a number?

What about a string and a string?

```python
# Which of the operators +, -, *, and / do something useful with the lists [1,10, 100] and [5, 4, 7]?

list_a = [1,10, 100]
list_b = [5, 4, 7]

list_final = list_a + list_b
print(f''' The + operator is beneficial for concatenating two lists while the others operators are not useful in this situation. \n The final list is '{list_final}' ''')

# What happens if you apply the operators +, -, *, / to a list and a number?
first_list = [1, 2, 3, 4]
second_list = 10
final_list = first_list * second_list
print()
print(f'''The * operator is beneficial for repeating the elements of the list a specified number of times while the others operators are not useful. The list in this situation is \n '{sorted(final_list)}' ''')

# What about a string and a string?
first_string = 'Christmas'
second_string = 'holiday'
final_string = str(first_string) + ' ' + str(second_string)
print()
print(f'''The + operator is beneficial for concatenating two strings while the others operators are not useful. \n The final string is '{final_string}' ''')
```

Work with a partner to design a data structure to represent a maze using dictionaries and lists.
- Each place in the maze has a name, which is a string.
- Each place in the maze has one or more people currently standing at it, by name.
- Each place in the maze has a maximum capacity of people that can fit in it.
- From each place in the maze, you can go from that place to a few other places, using a direction like 'up', 'north', or 'sideways'

Create an example instance, in a notebook, of a simple structure for your maze:
- The front room can hold 2 people. James is currently there. You can go outside to the garden, or upstairs to the bedroom, or north to the kitchen.
- From the kitchen, you can go south to the front room. It fits 1 person.
- From the garden you can go inside to front room. It fits 3 people. Sue is currently there.
- From the bedroom, you can go downstairs to the front room. You can also jump out of the window to the garden. It fits 2 people.

Make sure that your model:
- Allows empty rooms
- Allows you to jump out of the upstairs window, but not to fly back up.
- Allows rooms which people can't fit in.

```python
house = {
    'theFrontRoom': {'directionsToRoom': {'to go': 'to the garden',
                                          'upstairs': 'bedroom',
                                          'north': 'kitchen'},
                     'capacity': 2,
                     'peopleInRoom': []
                    },

    'kitchen': {'directionsToRoom': {'south': 'the front room'},
                'capacity': 1,
                'peopleInRoom': ['James']
                },

    'garden': {'directionsToRoom': {'inside': 'the front room'},
               'capacity': 2,
               'peopleInRoom': ['Sue']
               },

    'bedroom': {'directionsToRoom': {'downstairs': 'the front room', 'jump': 'garden'},
                'capacity': 2,
                'peopleInRoom': []
                }

}
```

Take your maze data structure. Write a program to count the total number of people in the maze, and also determine the total possible occupants.

```python

total_possible_ocupatans = sum(
    room["capacity"] for key, room in house.items() if "capacity" in room
    )

total_number_of_people = sum(
    len(room["peopleInRoom"]) for key, room in house.items() if "peopleInRoom" in room
    )

print(f'''The maximum occupancy of the house is {total_possible_ocupatans}, and currently, there are {total_number_of_people} people already present.''')
```
